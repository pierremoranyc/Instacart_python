```python
# Importing Dataframe
ords_prods_merged = pd.read_pickle(os.path.join(path, 'Data', 'Prepared Data', 'merged_orders_products_combi.pkl'))
```

```python
# Creating a Subset to avoid memory error
subset_ords_prods = ords_prods_merged[:1000000]
```

```python
# Defining a new function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# creating a new column 
subset_ords_prods['price_range']= subset_ords_prods.apply(price_label, axis=1)
```

```python
#checking values on the new columun
subset_ords_prods['price_range'].value_counts(dropna = False)
```

# Busiests hours of day

```python
result = []

for value in ords_prods_merged["order_hour_of_day"]:
    if value in (10, 11, 14, 15, 13, 12, 16, 9):
        result.append("Most Orders")
    elif value in (17, 8, 18, 19, 20, 7, 21, 22):
        result.append("Average Orders")
    elif value in (23, 6, 0, 1, 5, 2, 4, 3):
        result.append("Lowest Orders")

ords_prods_merged["busiest_period_of_day"] = result

```
