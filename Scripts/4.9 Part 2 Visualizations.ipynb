```python
#creating a line chart (prices & hour_of_day)

# Creating a small sample of the dataframe
np.random.seed(4)
dev = np.random.rand(len(ords_prods_cust_merged)) <= 0.7
```

```python
# store 70% of the sample in a Big dataframe
big = ords_prods_cust_merged[dev]
```

```python
# store 30% of the sample in a small dataframe
small = ords_prods_cust_merged[~dev]
```

```python
# checking accuracy 
len(ords_prods_cust_merged)
```

```python
len(big)+ len(small) 
```

```python
# creating a new dataframe using the 30% accurate with only the necessary columns for the line chart
df_2 = small[['order_hour_of_day','prices']]
```

```python
# creating a line chart with the new dataframe
line_2 = sns.lineplot(data = df_2, x = 'order_hour_of_day',y = 'prices')
```

Possible explanations of large error bands:  Data Volume: Before 10 am, there might be fewer orders and, consequently, a smaller volume of data points in the sample. When we have a smaller sample size, the variability in the data can lead to larger error bands. As we move beyond 10 am, the sample size likely increases, making the estimate of the price more stable, resulting in smaller error bands.
Order Patterns: Customer order patterns may change throughout the day. Before 10 am, there might be sporadic or infrequent orders, leading to larger fluctuations in prices due to less data to smooth out the variations. After 10 am, when orders become more frequent, the average price stabilizes, resulting in smaller error bands.

```python

# Create a line chart using Seaborn
plt.figure(figsize=(10, 6))
line_2 = sns.lineplot(data=df_2, x='order_hour_of_day', y='prices')
plt.title('Prices Variation During the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Price')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add legend
plt.legend(labels=['Prices'], loc='upper right')

# Save the line chart as a PNG image
plt.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'line_chart_prices_variation.png'), dpi=300, bbox_inches='tight')

# Show the line chart
plt.show()
```
