 # Task 4.8 (7)
 Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column

```python
#Missing values were changed to '0', but after checking the dataframe results, the ceros will affect the median_order_day observations, missing values will return to the original value of NaN.



ords_prods_merged.loc[ords_prods_merged['days_since_prior_order'] == 0, 'days_since_prior_order'] = np.nan

```

```python
ords_prods_merged['median_order_day'] = ords_prods_merged.groupby(('user_id'))['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merged.head(20)
```

# Flag criteria using loc() function: 

```python
ords_prods_merged.loc[ords_prods_merged['median_order_day'] > 20 , 'frequency_flag'] = 'Non_frequent_customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['median_order_day'] > 10) & (ords_prods_merged['median_order_day'] <= 20) , 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['median_order_day'] <= 10, 'frequency_flag']= 'frequent_customer'
```

```python
ords_prods_merged['frequency_flag'].value_counts(dropna= False)
```
